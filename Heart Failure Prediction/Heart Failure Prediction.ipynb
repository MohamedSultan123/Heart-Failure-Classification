```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
print(X.shape)
print(y.shape)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0,shuffle=True)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(copy=True, feature_range=(10, 100))
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
print(X_train)
print(X_test)
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
clss = LogisticRegression(penalty='l2', solver='sag', C=1, max_iter=500, random_state=33, tol=0.0001)
clss.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = clss.predict(X_test)
y_pred 
```

```python
print(clss.n_iter_)
print(clss.classes_)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import zero_one_loss
```

```python

#Calculating Confusion Matrix
CM = confusion_matrix(y_test, y_pred)
print('Confusion Matrix is : \n', CM)

# drawing confusion matrix
sns.heatmap(CM, center = True)
plt.show()
```

```python
y_train_pred = clss.predict(X_train)
train_acc = accuracy_score(y_train, y_train_pred)
print('Training Accuracy Score is :', train_acc)

#Calculating Accuracy Score  : ((TP + TN) / float(TP + TN + FP + FN))
AccScore = accuracy_score(y_test, y_pred)
print('Accuracy Score for test is : ', AccScore)
```

```python
ZeroOneLossValue = zero_one_loss(y_test,y_pred,normalize=False) 
print('Zero One Loss Value : ', ZeroOneLossValue )
```
